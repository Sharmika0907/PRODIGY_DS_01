```python
import seaborn as sns
import pandas as pd
titanic = sns.load_dataset('titanic')

print(titanic.head())
print(titanic['age'].describe())
print(titanic['sex'].value_counts())

import matplotlib.pyplot as plt
titanic['sex'].value_counts().plot(kind='bar', color=['skyblue', 'lightcoral'])

plt.title('Distribution of Genders on Titanic')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()


```
